```
import random
import time

class BNode:
    def __init__(self, item):
        self.item = item
        self.left = None
        self.right = None

    def setLeft(self, node):
        self.left = node

    def setRight(self, node):
        self.right = node

class BSTree:
    def __init__(self, root):
        self.root = root

    def insert(self, item):
        self.root = self._insert(self.root, item)

    def _insert(self, curNode, item):
        if curNode == None:
            curNode = BNode(item)
        elif item < curNode.item:
            curNode.left = self._insert(curNode.left, item)
        else:
            curNode.right = self._insert(curNode.right, item)
        return curNode

    def search(self, item):
        return self._search(self.root, item)

    def _search(self, curNode, item):
        if curNode == None:
            print("item not found")
        elif curNode.item == item:
            return curNode
        elif curNode.item > item:
            return self._search(curNode.left, item)
        elif curNode.item < item:
            return self._search(curNode.right, item)

    def inOrder(self):
        return self._inOrder(self.root, [])

    def _inOrder(self, node, sorted=[]):
        if node is not None:
            self._inOrder(node.left, sorted)
            sorted.append(node.item)
            self._inOrder(node.right, sorted)
        return sorted

    def minimum(self):
        return self._minimum(self.root.left, self.root)

    def _minimum(self, node):
        if node.left != None:
            return self._minimum(node.left)
        else:
            return node

    def delMin(self):
        self.root = self._delMin(self.root)

    def _delMin(self, node):
        if node.left == None:
            return node.right
        node.left = self._delMin(node.left)
        return node

    def delete(self, item):
        self.root = self._delete(self.root, item)

    def _delete(self, node, item):
        if node == None:
            return None
        if node.item > item:
            node.left = self._delete(node.left, item)
        elif node.item < item:
            node.right = self._delete(node.right, item)
        else:
            if node.right == None:
                return node.left
            elif node.left == None:
                return node.right
            delNode = node
            node =  self._minimum(delNode.right)
            node.right = self._delMin(delNode.right)
            node.left = delNode.left
        return node

data = random.sample(list(range(1, 1001)), 100)

root = BNode(data[0])
datatree = BSTree(root)


for i in range(1, 100):
  datatree.insert(data[i])

for i in range(100):
    # 100번 테스트 후, 평균
    target = random.sample(data, 1)
    summ1 = summ2 = 0
    
    listTime = time.time()
    for j in range(len(data)):
        if data[j] == target[0]:
            break
    summ1 += time.time() - listTime

    treeTime = time.time()
    datatree.search(target[0])
    summ2 += time.time() - treeTime

list_avg = summ1 / 100
tree_avg = summ2 / 100

print(list_avg, tree_avg)
print(list_avg - tree_avg)
```
