```
class BinaryTree:
    def __init__(self):
        self.t = [None]

    def append(self, item):
        self.t.append(item)

    def size(self):
        return len(self.t) - 1

    def getChild(self, item):
        if item in self.t:
            k = self.t.index(item)
            lidx = 2 * k
            ridx = 2 * k + 1
            if lidx <= self.size():
                lnode = self.t[lidx]
            else:
                lnode = None
            if ridx <= self.size():
                rnode = self.t[ridx]
            else:
                rnode = None
            return lnode, rnode
        else:
            print('item not found~')

    def getParent(self, item):
        if item in self.t:
            k = self.t.index(item)
            pidx = k // 2
            if pidx > 0:
                return self.t[pidx]
            else:
                return None
        else:
            print('item not found~')

    def cal(self, item1, item2):
        if self.getParent(item1) == self.getParent(item2):
            print(1/3)
        elif self.getParent().getParent(item1) == self.getParent().getParent(item2):
            print(2/3)
        elif self.getParent().getParent().getParent(item1) == self.getParent().getParent().getParent(item2):
            print(3/3)

tree = BinaryTree()

tree.append("호흡기/소화기병")
tree.append("호흡기병")
tree.append("소화기병")
tree.append("호흡기감영")
tree.append("폐질환")
tree.append("위질환")
tree.append("결장질환")
tree.append("독감")
tree.append("기관지염")
tree.append("폐부종")
tree.append("폐색전증")
tree.append("위궤양")
tree.append("위암")
tree.append("대장염")
tree.append("대장암")

tree.cal("위궤양", "위암")
```
