```
class minHeap:
    def __init__(self):
        self.x = [None]

    def _exchange(self, i, j):
        if self.x[i] > self.x[j]:
            self.x[i], self.x[j] = self.x[j], self.x[i]

    def push(self, item):
        self.x.append(item)
        cIndex = len(self.x) - 1
        pIndex = cIndex // 2
        while cIndex > 1:
            self._exchange(pIndex, cIndex)
            cIndex = pIndex
            pIndex = cIndex // 2

    def pop(self):
        item = self.x[1]
        n = len(self.x) - 1
        self.x[1], self.x[n] = self.x[n], self.x[1]
        del self.x[n]
        self.heapify()
        return item

    def _child(self, pIndex):
        n = len(self.x)
        leftIndex = pIndex * 2
        rightIndex = pIndex * 2 + 1
        if rightIndex <= n - 1:
            return leftIndex, rightIndex
        elif leftIndex == n - 1:
            return leftIndex, None
        else:
            return None, None

    def heapify(self):
        pIndex = 1
        lastIndex = len(self.x) - 1
        while pIndex < len(self.x):
            left, right = self._child(pIndex)
            if left != None and right == None:
                self._exchange(pIndex, left)
            elif left != None and right != None:
                if self.x[left] > self.x[right]:
                    self._exchange(pIndex, right)
                else:
                    self._exchange(pIndex, left)
            pIndex += 1

    def print(self):
        print(self.x)

class Item():
    def __init__(self, price):
        self.price = price

class Shop():
    def __init__(self):
        self.h = minHeap()

    def entItem(self, Item):
        self.h.push(Item)

    def outItem(self):
        money = self.h.pop()
        return money

    def size(self):
        return len(self.h.x) - 1

    def print(self):
        self.h.print()

price = np.random.normal(10000, 1000, 100)
price = price.tolist()
price = [int(value) for value in price]
benefit = []

s = Shop()

for i in range(100):
    mouse = Item(price[i])
    if s.size() < 9:
        s.entItem(mouse.price)
    else:
        s.entItem(mouse.price)
        benefit.append(s.outItem() - 7000)


print(benefit)

summ = 0
for i in benefit:
    summ += i

average = summ / len(benefit)
print(average)
```
