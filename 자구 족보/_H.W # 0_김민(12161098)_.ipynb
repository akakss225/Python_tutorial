# <H.W # 0> 이진수를 10진수로 바꾸는 프로그램을 작성하시오.

```
# 이진수를 10진수로 바꾸는 프로그램
bin = input("10진수로 바꾸고자 하는 이진수를 입력하시오 : ")
bin = int(bin)
dec = []

while True:
  fraction = bin % 10
  dec.append(fraction)
  bin = bin // 10
  if bin == 0: break

dec.reverse()
num = 0

for i in range(0, len(dec)):
  num = num + dec[i] * 2 ** i
print(num)
```
